## EDA

```python
import pandas as pd

from sklearn.decomposition import PCA

import numpy as np

from sklearn.preprocessing import scale 

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
#importing dataset
data = pd.read_csv('wine.csv')
data.head()
```

```python
data.tail()
```

```python
#identifying duplicates
data[data.duplicated()]
```

```python
#identifying null values
data.isna().sum()
```

```python
#Summary of data
data.info()
```

```python
data.describe()
```

## Visualization

```python
#plotting a pairplot to get visual insights
sns.pairplot(data,hue='Type')
```

```python
#dropping first coloumn
data = data.iloc[:,1:]
```

```python
data.head()
```

```python
#converting the data into array format

datascaled = data.values
datascaled
```

```python
#normalizing the data as there is scaling difference
datanorm = scale(datascaled)
datanorm
```

## PCA implementation

```python
pca = PCA()
pca_values = pca.fit_transform(datanorm)
pca_values
```

```python
# PCA components Matrix
pca.components_
```

```python
#variance of each PCA
var = pca.explained_variance_ratio_
var
```

```python
#Cumulative variance of each PCA
var1 = np.cumsum((np.round(var,4))*100)
```

```python
var1
```

```python
plt.plot(var1)
```

```python
#creating dataset for first three pc's
df = pd.DataFrame(pca_values[:,0:3],columns =['PC1','PC2','PC3'])
df
```

```python
wine = pd.read_csv('wine.csv')
wine
```

```python
final_df = pd.concat([wine['Type'],df],axis=1)
final_df
```

# Visualization of PC's

```python
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
fig = plt.figure(figsize=(15,10))
sns.scatterplot(data =final_df)
```

# Checking with other Clustering algorithms

#### Hierarchichal custering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
# create dendrogram
fig = plt.figure(figsize = (15,10))
dendrogram = sch.dendrogram(sch.linkage(datanorm,'complete'))
```

```python
#creating clusters
hc = AgglomerativeClustering(n_clusters = 3,affinity ='euclidean',linkage='ward')
hc
```

```python
hc.fit_predict(datanorm)
```

```python
#adding the clusters to the dataset
wine['Clusters'] = hc.labels_
wine
```

```python
wine['Clusters'].value_counts()
```

```python
#plotting clusters
sns.countplot(x='Clusters',data = wine)
plt.show()
```

### KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
WCSS=[]
for i in range(1,6):
        kmeans = KMeans(n_clusters = i,random_state=0)
        kmeans.fit(datanorm)
        WCSS.append(kmeans.inertia_)
        
plt.plot(range(1,6),WCSS,color='red')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Building CLustering algorithm
# k = 3
clusters_new = KMeans(3,random_state=42)
clusters_new.fit(datanorm)
```

```python
clusters_new.labels_
```

```python
wine2 = pd.read_csv('wine.csv')
wine2['ClustersKM']= clusters_new.labels_
wine2
```

```python
wine2['ClustersKM'].value_counts()
```

```python
# Ploting Clusters
plt.figure(figsize=(10, 7))
sns.countplot(x ='ClustersKM',data = wine2)
plt.show()
```
